```python
from pinecone import Pinecone, ServerlessSpec
from google.colab import userdata

pc = Pinecone(api_key=userdata.get('PINECONE_API_KEY'))
```

```python
index_name = "standard-dense-py"

if not pc.has_index(index_name):
    pc.create_index(
        name=index_name,
        vector_type="dense",
        dimension=768,
        metric="cosine",
        spec=ServerlessSpec(
            cloud="aws",
            region="us-east-1"
        ),
        deletion_protection="disabled",
        tags={
            "environment": "development"
        }
    )
```

```python
# Target the index
dense_index = pc.Index(index_name)
```

```python
import uuid

# Prepare vectors for standard upsert (not upsert_records)
vectors = []
for i in range(len(text_chunk)):
    vectors.append({
        "id": str(uuid.uuid4()),
        "values": embedding.embed_query(text_chunk[i].page_content),
        "metadata": {
            "chunk_text": text_chunk[i].page_content
        }
    })

print(f"Prepared {len(vectors)} vectors for upsert")
```

```python
batch_size = 96
for i in range(0, len(vectors), batch_size):
    batch = vectors[i:i + batch_size]
    dense_index.upsert(
        vectors=batch,
        namespace="example-namespace"
    )
    print(f"Upserted batch {i//batch_size + 1}/{(len(vectors) + batch_size - 1)//batch_size}")

print("All vectors upserted successfully!")
```

```python
index = pc.Index(host='standard-dense-py-a3n7u7v.svc.aped-4627-b74a.pinecone.io')

index.query(
    namespace="example-namespace",
    vector=embedding.embed_query("What is command to create branch in git?"),
    top_k=3,
    include_metadata=True,
    include_values=False
)
```
