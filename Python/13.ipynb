### Question:
You have sales data stored in a dictionary, where each key is a product name and its value is a list of daily sales. Write Python code to calculate the total sales for each product, sort the products by their total sales in descending order, and plot the results using matplotlib.

```python
import matplotlib.pyplot as plt
```

```python
# Sales data
sales_data = {
    "Product A": [120, 130, 150, 110],
    "Product B": [90, 100, 80, 70],
    "Product C": [200, 220, 210, 250],
    "Product D": [50, 60, 55, 45]
}
```

```python
# Calculate total sales for each product
total_sales = {product: sum(sales) for product, sales in sales_data.items()}
```

```python
# Sort products by total sales in descending order
sorted_sales = sorted(total_sales.items(), key=lambda x: x[1], reverse=True)
```

```python
# Unpack sorted data
products, totals = zip(*sorted_sales)
```

```python
# Plot the results
plt.figure(figsize=(8, 6))
plt.bar(products, totals, color="skyblue", edgecolor="black")
plt.title("Total Sales by Product", fontsize=16)
plt.xlabel("Product", fontsize=14)
plt.ylabel("Total Sales", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
