### Question:
Write a Python script to analyze a dataset using an inbuilt dataset from sklearn. Use the load_iris dataset to:

1. Display the first five rows of the dataset.
2. Perform basic statistics like mean, median, and standard deviation for each feature.
3. Create a scatter plot to visualize the relationship between two chosen features (e.g., sepal length and sepal width).

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = load_iris()
# Convert to a pandas DataFrame for easier manipulation
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
# 1. Display the first five rows of the dataset
print("First five rows of the dataset:")
print(df.head())
```

```python
# 2. Perform basic statistics
print("\nBasic Statistics for Each Feature:")
for column in df.columns:
    print(f"\nFeature: {column}")
    print(f"Mean: {df[column].mean():.2f}")
    print(f"Median: {df[column].median():.2f}")
    print(f"Standard Deviation: {df[column].std():.2f}")
```

```python
# 3. Create a scatter plot for sepal length and sepal width
plt.figure(figsize=(8, 6))
plt.scatter(df[iris.feature_names[0]], df[iris.feature_names[1]], c='blue', alpha=0.6)
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)
plt.show()
```
