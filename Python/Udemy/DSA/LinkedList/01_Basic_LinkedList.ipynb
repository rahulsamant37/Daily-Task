# Linked List
A linked list is a type of data structure in which elements are connected to each other through links or pointers. Each element (called a node) has two parts - the data and a link pointing to the next node.



### 1. Brute Force Approch

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

    
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)

node1.next = node2
node2.next = node3

head = node1

print(head.next.value)
print(head.next.next.value)
# print(head.next.next.next.value) # AttributeError: 'NoneType' object has no attribute 'value' out of range
```

### 2. Optimal Approach for triggering every element

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def print_LL(head):
    while(head!=None):
        print(head.value)
        head = head.next
    return
    
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)

print(print_LL(node1))
```

```python
## Return a head to a newly created LL
def take_input():
    value = int(input("Enter the value of Node:- "))
    head = None

    while(value != -1):
        newNode = Node(value)
        if head==Node:
            head = newNode
        else:
            head.next = newNode
        
        value = int(input("Enter the value of Node:- "))
    return head

new_head = take_input()
print_LL(new_head)
```
