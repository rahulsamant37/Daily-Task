## PyTorch in One Hour: From Tensors to Training Neural Networks on Multiple GPUs

```python
import torch

# Create a 0D tensor
tensor0D = torch.tensor(1)

# Create a 1D tensor
tensor1D = torch.tensor([1, 2, 3])

# Create a 2D tensor
tensor2D = torch.tensor([[1, 2], [3, 4]])

# Create a 3D tensor
tensor3D = torch.tensor([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
```

#### NOTE: PyTorch adopts the default 64-bit integer data type from Python

```python
print(tensor1D.dtype)
```

#### If we create tensors from Python floats, PyTorch creates tensors with a 32-bit precision by default, as we can see below:

```python
floatvec = torch.tensor([2.0, 3.0, 4.0])
print(floatvec)
print(floatvec.dtype)
```

#### changing a 64-bit integer tensor into a 32-bit float tensor

```python
floatvec = tensor1D.to(torch.float32)
print(floatvec)
print(floatvec.dtype)
```

```python
tensor2D = torch.tensor([[1, 2, 3], [4, 5, 6]])
tensor2D
```

```python
# shape of the tensor
print(tensor2D.shape)
```

#### For reshaping the tensor dimensions

##### Method 01: use `.reshape()`

```python
tensor2D.reshape(3, 2)
```

##### Method 02: use `.view()`

```python
tensor2D.view(2,3)
```

#### Transpose a tensor

```python
tensor2D.T
```

#### To multiply two matrices in Pytorch

##### Method 01: `.matmul` method:

```python
tensor2D.matmul(tensor2D.T)
```

##### Method 02: use `@` operator:

```python
tensor2D @ tensor2D.T
```
