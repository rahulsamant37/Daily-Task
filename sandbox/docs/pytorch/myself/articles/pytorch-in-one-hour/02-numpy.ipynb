```python
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def binary_cross_entropy(y, a):
    a = np.clip(a, 1e-15, 1 - 1e-15)
    return -np.mean(y*np.log(a) + (1-y) * np.log(1-a))
```

```python
y = np.array([1.0])
x1 = np.array([1.1])
w1 = np.array([2.2])
b = np.array([0.0])

z = x1 * w1 + b
a = sigmoid(z)

loss = binary_cross_entropy(y, a)
print(loss)
```
