```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer

from sdv.metadata import Metadata
from sdv.single_table import CTGANSynthesizer, TVAESynthesizer
```

```python
df = load_breast_cancer(as_frame=True).frame

train, test = train_test_split(df, test_size=0.2)

metadata = Metadata.detect_from_dataframe(train, table_name='breast_cancer')
metadata.update_column('target',table_name='breast_cancer', sdtype='categorical')

metadata.save_to_json('breast_cancer_metadata.json')

CTGANsynthesizer = CTGANSynthesizer(metadata, epochs=100)
CTGANsynthesizer.fit(train)
TVAEsynthesizer = TVAESynthesizer(metadata, epochs=100)
TVAEsynthesizer.fit(train)
```

```python
CTGANsynthetic = CTGANsynthesizer.sample(num_rows=len(train))
TVAEsynthetic = TVAEsynthesizer.sample(num_rows=len(train))
```
